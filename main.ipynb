```python
import json
import os

tarefas = []

def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def linha_separadora():
    print("=" * 60)

def cabecalho():
    limpar_tela()
    linha_separadora()
    print("ğŸ“‹ GERENCIADOR DE TAREFAS".center(60))
    linha_separadora()

def salvar_tarefas():
    with open("tarefas.json", "w") as arquivo:
        json.dump(tarefas, arquivo, indent=2)
    print("\nâœ… Tarefas salvas com sucesso!")
    input("\nPressione ENTER para continuar...")

def carregar_tarefas():
    try:
        with open("tarefas.json", "r") as arquivo:
            tarefas = json.load(arquivo)
            print("âœ… Tarefas carregadas com sucesso!")
            return tarefas
    except:
        print("âš ï¸  Nenhum arquivo encontrado! ComeÃ§ando uma nova lista.")
        return []

def adicionar_tarefa(nome, prioridade):
    tarefa = {
        "nome": nome,
        "prioridade": prioridade
    }

    tarefas.append(tarefa)
    print(f"\nâœ… Tarefa adicionada com sucesso!")
    print(f"ğŸ“ {tarefa['nome']} | ğŸ”¥ Prioridade: {tarefa['prioridade']}")
    input("\nPressione ENTER para continuar...")

def listar_tarefas():
    cabecalho()
    print("ğŸ“‹ LISTA DE TAREFAS")
    linha_separadora()

    if not tarefas:
        print("ğŸ“­ Nenhuma tarefa encontrada!")
    else:
        for i, tarefa in enumerate(tarefas, 1):
            prioridade_icon = "ğŸ”´" if tarefa['prioridade'].lower() in ['alta', 'urgente'] else "ğŸŸ¡" if tarefa['prioridade'].lower() == 'mÃ©dia' else "ğŸŸ¢"
            print(f"{i:2d}. {prioridade_icon} {tarefa['nome']}")
            print(f"     â””â”€ Prioridade: {tarefa['prioridade']}")
            print()

    linha_separadora()
    input("Pressione ENTER para voltar ao menu...")

def remover_tarefa(nome):
    for tarefa in tarefas:
        if tarefa["nome"].lower() == nome.lower():
            tarefas.remove(tarefa)
            print(f"\nâœ… Tarefa '{nome}' removida com sucesso!")
            input("\nPressione ENTER para continuar...")
            return

    print(f"\nâŒ Tarefa '{nome}' nÃ£o encontrada!")
    input("\nPressione ENTER para continuar...")

def mostrar_menu():
    cabecalho()
    print("ğŸ¯ MENU DE OPÃ‡Ã•ES")
    linha_separadora()
    print("1ï¸âƒ£  Adicionar Tarefa")
    print("2ï¸âƒ£  Listar Tarefas")
    print("3ï¸âƒ£  Remover Tarefa")
    print("4ï¸âƒ£  Salvar Tarefas")
    print("5ï¸âƒ£  Sair")
    linha_separadora()

# InicializaÃ§Ã£o
print("ğŸš€ Iniciando Gerenciador de Tarefas...")
tarefas = carregar_tarefas()
input("\nPressione ENTER para continuar...")

while True:
    mostrar_menu()

    try:
        opcao = int(input("ğŸ”¹ Digite uma opÃ§Ã£o (1-5): "))
    except ValueError:
        print("\nâŒ Por favor, digite apenas nÃºmeros!")
        input("Pressione ENTER para continuar...")
        continue

    if opcao == 1:
        cabecalho()
        print("â• ADICIONAR NOVA TAREFA")
        linha_separadora()
        nome = input("ğŸ“ Digite o nome da tarefa: ").strip()

        if not nome:
            print("\nâŒ Nome da tarefa nÃ£o pode estar vazio!")
            input("Pressione ENTER para continuar...")
            continue

        print("\nğŸ”¥ OpÃ§Ãµes de prioridade:")
        print("   â€¢ Alta/Urgente")
        print("   â€¢ MÃ©dia")
        print("   â€¢ Baixa")
        prioridade = input("\nğŸ¯ Digite a prioridade da tarefa: ").strip()

        if not prioridade:
            prioridade = "MÃ©dia"

        adicionar_tarefa(nome, prioridade)

    elif opcao == 2:
        listar_tarefas()

    elif opcao == 3:
        cabecalho()
        print("ğŸ—‘ï¸  REMOVER TAREFA")
        linha_separadora()

        if not tarefas:
            print("ğŸ“­ Nenhuma tarefa para remover!")
            input("Pressione ENTER para continuar...")
            continue

        print("ğŸ“‹ Tarefas disponÃ­veis:")
        for i, tarefa in enumerate(tarefas, 1):
            print(f"   {i}. {tarefa['nome']}")
        print()

        nome = input("ğŸ”¹ Digite o nome da tarefa a ser removida: ").strip()
        if nome:
            remover_tarefa(nome)
        else:
            print("\nâŒ Nome nÃ£o pode estar vazio!")
            input("Pressione ENTER para continuar...")

    elif opcao == 4:
        cabecalho()
        print("ğŸ’¾ SALVANDO TAREFAS...")
        linha_separadora()
        salvar_tarefas()

    elif opcao == 5:
        cabecalho()
        print("ğŸ‘‹ ENCERRANDO PROGRAMA")
        linha_separadora()
        print("Obrigado por usar o Gerenciador de Tarefas!")
        print("AtÃ© logo! ğŸš€")
        break

    else:
        print(f"\nâŒ OpÃ§Ã£o '{opcao}' invÃ¡lida! Escolha entre 1 e 5.")
        input("Pressione ENTER para continuar...")
```
